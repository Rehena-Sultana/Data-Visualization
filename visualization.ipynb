```python
import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
df=pd.read_csv('tips.csv')
df[:3]
```

Normally, when we try to solve real-time data problems, we have numerical and categorical data. In this analysis, I will try to showcase the possible ways of representing numerical and categorical plots.

# **Scatter plot for numerical values:**

we can implement scatterplot in 2 ways using seaborn:
   * 1) relplot
   * 2)scatterplot



```python
sns.relplot(x='total_bill',y='tip',data=df)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=df)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=df)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=df)
```

# **scatter plot for categorical column**



* stripplot
* swarmplot

```python
sns.catplot(x='sex',y='tip',kind='strip',jitter=0,data=df)
```

```python
#swarm plot
sns.catplot(x='sex',y='tip',kind='swarm',data=df)
```

```python
sns.catplot(x='sex',y='tip',kind='swarm',hue='smoker',data=df)
```

```python
sns.swarmplot(x='sex',y='tip',data=df)
```

A swarmplot differs from a stripplot in that they both provide the same information, but a swarmplot also includes the distribution concept.

# **Jointplot**<br>

# **Boxplot**

Box plot gives us a summary of data distribution with 5 tuples such as-minimum,1st quartile,median,3rd quartile,maximun.

* Q1= 25% quartile
* Q3= 75% QUARTILE

* IQR(Inter Quartile Range)=Q3-Q1

* minimun=Q1-1.5*IQR

* maximum=Q3+1.5*IQR

* The values less than the minimun and the values greater than the maximun are called outliers.


```python
sns.boxplot(df['tip'])
```

```python
sns.catplot(x='day',y='tip',data=df,kind='box')
```

# **PDF(Probability Density Function)**

```python
import scipy.stats as stats
df.sort_values(by=['tip'], inplace=True)
df_mean = np.mean(df["tip"])
df_std = np.std(df["tip"])
pdf = stats.norm.pdf(df["tip"], df_mean, df_std)

plt.plot(df["tip"], pdf)
```

```python
sns.distplot(df['tip'])
```
